```python
import json
f = open("./chicago-hotels.json","r")
data = json.load(f)
data
```

```python
new_list = sorted(data, key=lambda x: x['rating'], reverse=True)
for item in new_list:
    print(item['name'][:40], item['rating'])
```

```python
new_list = sorted(data, key=lambda x:(x['reviews'], x['rating']), reverse=True)
for item in new_list:
    print(item['name'][:40], item['reviews'],item['rating'])
```
